```python
def ransom_note(magazine, ransom):
    result = 'No'
    rc = {} # dict of word: count of that word in the note
    for word in note:
        if word not in rc:
            rc[word] = 0
        rc[word] += 1
    
    for word in magazine:
        if word in rc:
            rc[word] -= 1
            if rc[word] == 0:
                del rc[word]
                if not rc:
                    result = 'Yes'
    print (result)
        
    
```
